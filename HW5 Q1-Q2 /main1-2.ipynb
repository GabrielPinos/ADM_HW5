```python
# Filter significant routes
filtered_routes = df_clean[
    (df_clean['Flights'] >= 50) & 
    (df_clean['Seats'] >= 100)
]

# Calculate averages and statistics
route_avg = filtered_routes.groupby(['Origin_airport', 'Destination_airport']).agg(
    Total_Passengers=('Passengers', 'sum'),
    Total_Flights=('Flights', 'sum'),
    Total_Seats=('Seats', 'sum')
).reset_index()

# Calculate metrics
route_avg['Avg_Passengers_per_Flight'] = route_avg['Total_Passengers'] / route_avg['Total_Flights']
route_avg['Seat_Utilization'] = route_avg['Total_Passengers'] / route_avg['Total_Seats']

# Filter routes based on minimum seat utilization
route_avg = route_avg[route_avg['Seat_Utilization'] >= 0.25]

# Calculate global statistics
global_avg_passengers = route_avg['Avg_Passengers_per_Flight'].mean()
global_seat_utilization = route_avg['Seat_Utilization'].mean()

# Print statistics
print(f"Global Average Passengers per Flight: {global_avg_passengers:.2f}")
print(f"Global Seat Utilization: {global_seat_utilization:.4f}")

# Classify routes
route_avg['Usage_Category'] = route_avg['Seat_Utilization'].apply(
    lambda x: 'Under-Utilized' if x < global_seat_utilization else 'Over-Utilized'
)

# Top 10 over-utilized routes
top_10_over_utilized = route_avg[
    route_avg['Usage_Category'] == 'Over-Utilized'
].sort_values(by='Seat_Utilization', ascending=False).head(10)

# Top 10 under-utilized routes
top_10_under_utilized = route_avg[
    route_avg['Usage_Category'] == 'Under-Utilized'
].sort_values(by='Seat_Utilization', ascending=True).head(10)

# Top 10 routes with the highest average passengers per flight
top_10_highest_avg_passengers = route_avg.sort_values(by='Avg_Passengers_per_Flight', ascending=False).head(10)

# Display the results
print("\nTop 10 Over-Utilized Routes:")
display(top_10_over_utilized[['Origin_airport', 'Destination_airport', 'Seat_Utilization', 'Total_Passengers', 'Total_Flights']])

print("\nTop 10 Under-Utilized Routes:")
display(top_10_under_utilized[['Origin_airport', 'Destination_airport', 'Seat_Utilization', 'Total_Passengers', 'Total_Flights']])

print("\nTop 10 Routes with the Highest Average Passengers per Flight:")
display(top_10_highest_avg_passengers[['Origin_airport', 'Destination_airport', 'Avg_Passengers_per_Flight', 'Total_Passengers', 'Total_Flights']])
```

# Alternative Centrality Measures Suggested by the LLM

The LLM suggested using Eigenvector Centrality, which measures the importance of a node by considering not only its number of connections but also the importance of the nodes it connects to. This is particularly useful in transportation networks, where being connected to major hubs increases an airports significance.

Since the flight network is a directed graph, Eigenvector Centrality was adapted to consider only incoming edges. This customized version was implemented as LLM Centrality.

# How to Verify the Reliability of LLM Results

The primary way to verify the reliability of the LLMs suggested metric is to compare its results with those of the four centrality measures already computed: Degree Centrality, Closeness Centrality, Betweenness Centrality, and PageRank.

If the airports identified by the LLMs suggested metric also appear among the top-ranked airports in the traditional metrics or if they provide new information consistent with the networks structure, this indicates that the metric is reliable and useful for analysis.

Additionally, comparing the results helps determine whether the suggested metric offers a different or complementary perspective compared to the existing metrics, improving the overall understanding of the network.
