```python
def dijkstra_manual(graph_edges, source, target):
    nodes = set([edge[0] for edge in graph_edges] + [edge[1] for edge in graph_edges]) # Extract all unique nodes from the graph edges
    # Initialize distances to all nodes as infinity, except the source which is set to 0
    distances = {node: float('inf') for node in nodes}
    distances[source] = 0

    #Dictionary to store the previous node
    previous_nodes = {node: None for node in nodes}
    unvisited_nodes = nodes.copy()

    while unvisited_nodes: # Loop until all nodes are visited
        current_node = None
        min_distance = float('inf')

        # Find the unvisited node with the smallest distance
        for node in unvisited_nodes:
            if distances[node] < min_distance:
                min_distance = distances[node]
                current_node = node

        if current_node is None or current_node == target: # If no node is reachable or target is found exit the loop
            break

        unvisited_nodes.remove(current_node)
        neighbors = [edge for edge in graph_edges if edge[0] == current_node]# Find all neighbors of the current node

        # Update distances to neighbors if a shorter path is found
        for edge in neighbors:
            neighbor, weight = edge[1], edge[2]
            new_distance = distances[current_node] + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                previous_nodes[neighbor] = current_node

    # Backtrack to find the path from source to target
    path = []
    current_node = target
    while current_node is not None:
        path.insert(0, current_node) # Add the node to the beginning of the path
        current_node = previous_nodes[current_node] # Move to the previous node

     # If no valid path exists, return "No route found"
    if not path or path[0] != source:
        return "No route found", float('inf')

    return path, distances[target] # Return the shortest path and its total distance


def find_all_routes_and_best(df, origin_city, destination_city, flight_date):
    flights_on_date = df[df['Fly_date'] == flight_date] # Filter the dataframe for flights on the specified date
    origin_airports = flights_on_date[flights_on_date['Origin_city'] == origin_city]['Origin_airport'].unique() # Get all unique airports in the origin city
    destination_airports = flights_on_date[flights_on_date['Destination_city'] == destination_city]['Destination_airport'].unique() # Get all unique airports in the destination city

    graph_edges = flights_on_date[['Origin_airport', 'Destination_airport', 'Distance']].values.tolist() # Create a list of edges representing the graph
    results = []

    # Initialize variables to track the best route and its distance
    best_route = None
    best_distance = float('inf')

    # Iterate over all airport pairs between the origin and destination cities
    for origin_airport in origin_airports:
        for destination_airport in destination_airports:
            # Check if the origin and destination are connected in the graph
            if origin_airport in [edge[0] for edge in graph_edges] and destination_airport in [edge[1] for edge in graph_edges]:
                route, distance = dijkstra_manual(graph_edges, origin_airport, destination_airport) # Use Dijkstra's algorithm to find the shortest path and distance
                route_str = " â†’ ".join(route) if route != "No route found" else "No route found"
            else: # If no connection exists, set route and distance to default values
                route, distance = "No route found", float('inf')
                route_str = "No route found"

            results.append({
                "Origin_city_airport": origin_airport,
                "Destination_city_airport": destination_airport,
                "Best_route": route_str,
                "Distance": distance
            })

            # Update the best route if a shorter one is found
            if route != "No route found" and distance < best_distance:
                best_route = route_str
                best_distance = distance

    result_df = pd.DataFrame(results)

    if not result_df.empty and "Best_route" in result_df.columns:
        result_df['Is_Best'] = result_df['Best_route'] == best_route

    display(result_df)
    return result_df, best_route
```

The first code implements Dijkstraâ€™s algorithm to manually calculate the shortest path between two airports within a flight network graph. The dijkstra_manual function initializes nodes with distances set to infinity and iteratively updates these distances to find the minimum distance path. The find_all_routes_and_best function filters flights for a specific date and identifies all possible airport pairs for the origin and destination cities. For each pair, it calculates the optimal route using dijkstra_manual and selects the one with the shortest distance as the â€œbest route.â€ Finally, it creates a DataFrame with the results for each airport pair, highlighting the best route.

```python
import folium

def visualize_routes_on_map(df, routes, best_route, map_filename="all_routes_map.html"):
    
    # Initialize the map object as None, to be created later when coordinates are available
    m = None

    # Iterate over each route in the routes dataframe
    for route_data in routes.to_dict('records'):
        route = route_data['Best_route']
        if route == "No route found": # Skip routes without valid paths
            continue

        coordinates = []
        airport_sequence = route.split(" â†’ ")  # Decode the sequence "A â†’ B â†’ C"

        # Iterate over each airport in the route
        for airport in airport_sequence:
            row = df[(df['Origin_airport'] == airport) | (df['Destination_airport'] == airport)].iloc[0]

            # If the airport matches the origin add its coordinates
            if airport in row['Origin_airport']:
                coordinates.append((row['Org_airport_lat'], row['Org_airport_long'], airport))

            # If the airport matches the destination add its coordinates    
            elif airport in row['Destination_airport']:
                coordinates.append((row['Dest_airport_lat'], row['Dest_airport_long'], airport))

        #create the map at the first route's starting location
        if m is None:
            m = folium.Map(location=coordinates[0][:2], zoom_start=6)

        #green for the best route, red for others
        color = "green" if route == best_route else "red"
        for i in range(len(coordinates) - 1): # Add a line
            folium.PolyLine([coordinates[i][:2], coordinates[i + 1][:2]], color=color, weight=2.5, opacity=1).add_to(m)
        for coord in coordinates: # Add markers
            folium.Marker(location=coord[:2], popup=f"Airport: {coord[2]}").add_to(m)

    if m is not None:
        m.save(map_filename)
        print(f"Map of routes saved in '{map_filename}'.")
    else: # if no valid routes were found to display
        print("No valid routes to display.")
```

This code visualizes all the routes calculated by the previous code on an interactive map using the folium library. The visualize_routes_on_map function highlights the best route in green and all other routes in red. For each valid route, the geographical points (latitude and longitude) of each airport are determined, and the segments connecting the nodes are drawn. Additionally, the map displays markers for each airport with a popup showing the airport name. The resulting map is saved as an HTML file for easy viewing.

```python
def count_unique_routes_for_date(df, specific_date):
   
    flights_on_date = df[df['Fly_date'] == specific_date]

    # Group flights by origin city, destination city, origin airport, and destination airport, and count occurrences to calculate the number of flights per route
    route_counts = (
        flights_on_date.groupby(['Origin_city', 'Destination_city', 'Origin_airport', 'Destination_airport'])
        .size()
        .reset_index(name='Count')
    )

    # group by origin city and destination city to count unique routes between each city pair
    unique_routes = (
        route_counts.groupby(['Origin_city', 'Destination_city'])
        .size()
        .reset_index(name='Unique_Routes')
    )

    unique_routes_sorted = unique_routes.sort_values(by='Unique_Routes', ascending=False) #descending order

    return unique_routes_sorted

specific_date = "2000-11-01"

unique_routes_date_df = count_unique_routes_for_date(df_clean, specific_date)

print(f"Cities with the highest number of unique routes for the date {specific_date}:")
display(unique_routes_date_df)
```
