```python
# Importa le librerie necessarie
import pandas as pd

# Carica il dataset
file_path = "Airports2.csv"
df = pd.read_csv(file_path)
```

```python
# Rimuovi righe con valori mancanti nelle coordinate geografiche
df_clean = df.dropna(subset=['Org_airport_lat', 'Org_airport_long', 
                             'Dest_airport_lat', 'Dest_airport_long'])

# Verifica dei valori mancanti dopo la pulizia
missing_values_clean = df_clean.isnull().sum()
print("Valori mancanti dopo la pulizia:")
display(missing_values_clean[missing_values_clean > 0])

# Mostra il numero totale di righe rimanenti
print(f"Righe rimanenti dopo la pulizia: {len(df_clean)}")
```

```python
import networkx as nx

# Creazione del grafo non orientato
flight_graph = nx.Graph()

# Aggiunta degli archi con i pesi (numero di voli)
for _, row in df_clean.iterrows():
    flight_graph.add_edge(row['Origin_airport'], row['Destination_airport'], weight=row['Flights'])

# Numero di nodi e archi
num_nodes = flight_graph.number_of_nodes()
num_edges = flight_graph.number_of_edges()

print(f"Numero di nodi (aeroporti): {num_nodes}")
print(f"Numero di archi (rotte): {num_edges}")

# Calcolo della densità del grafo (usando la formula della traccia)
density = (2 * num_edges) / (num_nodes * (num_nodes - 1))
print(f"Densità del grafo: {density:.4f}")
```

```python
# Carica il dataset completo
file_path = "Airports2.csv"
df_full = pd.read_csv(file_path)

# Mostra le prime righe per conferma
print("Prime righe del dataset completo:")
display(df_full.head())
```

```python
# Rimuovi righe con 0 passeggeri o 0 voli
df_cleaned = df_full[(df_full['Passengers'] > 0) & (df_full['Flights'] > 0)]

# Mostra il numero di righe rimanenti
print(f"Righe rimanenti dopo la pulizia iniziale (0 passeggeri o voli rimossi): {len(df_cleaned)}")
```

```python
# Raggruppa e calcola il totale passeggeri e voli per ogni rotta
route_efficiency = df_cleaned.groupby(['Origin_airport', 'Destination_airport']).agg(
    Total_Passengers=('Passengers', 'sum'),
    Total_Flights=('Flights', 'sum')
).reset_index()

# Calcola la media dei passeggeri per volo
route_efficiency['Avg_Passengers_per_Flight'] = (
    route_efficiency['Total_Passengers'] / route_efficiency['Total_Flights']
)

# Mostra alcune righe per conferma
print("Prime righe della tabella con la media dei passeggeri per volo:")
display(route_efficiency.head())
```

```python
# Calcolo del 10° percentile
tenth_percentile = route_efficiency['Avg_Passengers_per_Flight'].quantile(0.10)
print(f"10° Percentile della Media Passeggeri per Volo: {tenth_percentile:.2f}")

# Filtro aggressivo: elimina rotte con una media sotto il 10° percentile
route_efficiency_filtered = route_efficiency[
    route_efficiency['Avg_Passengers_per_Flight'] >= tenth_percentile
].copy()  # Creiamo una copia esplicita per evitare il warning

# Mostra il numero di rotte rimanenti
print(f"Numero di rotte rimanenti dopo il filtro aggressivo: {len(route_efficiency_filtered)}")
```

```python
# Calcola la media globale aggiornata
global_avg_passengers_per_flight = route_efficiency_filtered['Avg_Passengers_per_Flight'].mean()
print(f"Media globale di passeggeri per volo (filtrata): {global_avg_passengers_per_flight:.2f}")

# Classifica le rotte sotto-utilizzate e sovra-utilizzate
route_efficiency_filtered['Usage_Category'] = route_efficiency_filtered['Avg_Passengers_per_Flight'].apply(
    lambda x: 'Sotto-Utilizzata' if x < global_avg_passengers_per_flight else 'Sovra-Utilizzata'
)
```

```python
# Ordinamento delle rotte sotto-utilizzate
under_utilized = route_efficiency_filtered[route_efficiency_filtered['Usage_Category'] == 'Sotto-Utilizzata'].sort_values(
    by='Avg_Passengers_per_Flight', ascending=True)

# Ordinamento delle rotte sovra-utilizzate
over_utilized = route_efficiency_filtered[route_efficiency_filtered['Usage_Category'] == 'Sovra-Utilizzata'].sort_values(
    by='Avg_Passengers_per_Flight', ascending=False)

# Mostra le prime 10 rotte sotto-utilizzate
print("Prime 10 rotte sotto-utilizzate (filtrate):")
display(under_utilized.head(10))

# Mostra le prime 10 rotte sovra-utilizzate
print("Prime 10 rotte sovra-utilizzate (filtrate):")
display(over_utilized.head(10))
```

```python
import matplotlib.pyplot as plt

# Istogramma della distribuzione filtrata
plt.figure(figsize=(8, 5))
plt.hist(route_efficiency_filtered['Avg_Passengers_per_Flight'], bins=50, color='skyblue', edgecolor='black')
plt.title("Distribuzione della Media dei Passeggeri per Volo (Filtrata)")
plt.xlabel("Media Passeggeri per Volo")
plt.ylabel("Frequenza")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd

# Carica il dataset completo
file_path = "Airports2.csv"
df_full = pd.read_csv(file_path)

# Filtro iniziale: rimuovi righe con 0 passeggeri, 0 voli o 0 posti a sedere
df_cleaned = df_full[(df_full['Passengers'] > 0) & (df_full['Flights'] > 0) & (df_full['Seats'] > 0)]
print(f"Righe rimanenti dopo la pulizia iniziale: {len(df_cleaned)}")

# Calcolo delle rotte più trafficate
route_efficiency = df_cleaned.groupby(['Origin_airport', 'Destination_airport']).agg(
    Total_Passengers=('Passengers', 'sum'),
    Total_Flights=('Flights', 'sum'),
    Total_Seats=('Seats', 'sum')
).reset_index()

# Calcola il rapporto tra passeggeri e posti a sedere per ogni rotta
route_efficiency['Seat_Utilization'] = (
    route_efficiency['Total_Passengers'] / route_efficiency['Total_Seats']
)

# Seleziona le top 1.000 rotte più trafficate
top_1000_routes = route_efficiency.sort_values(by='Total_Passengers', ascending=False).head(1000)

# Visualizza le prime 10 rotte più trafficate
print("Prime 10 rotte più trafficate (ordinato per Total_Passengers):")
display(top_1000_routes[['Origin_airport', 'Destination_airport', 'Total_Passengers', 'Total_Flights', 'Seat_Utilization']].head(10))

# Aggiungi le coordinate geografiche per ogni rotta
route_efficiency_geo = top_1000_routes.merge(
    df_full[['Origin_airport', 'Destination_airport', 
             'Org_airport_lat', 'Org_airport_long', 
             'Dest_airport_lat', 'Dest_airport_long']],
    on=['Origin_airport', 'Destination_airport'],
    how='left'
)

# Filtro aggiuntivo: rimuovi rotte con un utilizzo dei posti a sedere inferiore a una certa soglia
route_efficiency_geo_filtered = route_efficiency_geo[route_efficiency_geo['Seat_Utilization'] > 0.1]
print(f"Righe rimanenti dopo il filtro di utilizzo dei posti a sedere: {len(route_efficiency_geo_filtered)}")

# Rimuovi righe con valori nulli nelle coordinate
map_data = route_efficiency_geo_filtered.dropna(
    subset=['Org_airport_lat', 'Org_airport_long', 'Dest_airport_lat', 'Dest_airport_long']
)
print(f"Righe rimanenti dopo la rimozione di valori nulli: {len(map_data)}")

# Calcola la media del rapporto di utilizzo dei posti a sedere
global_seat_utilization = route_efficiency['Seat_Utilization'].mean()
print(f"Media globale di utilizzo dei posti a sedere per rotta: {global_seat_utilization:.2f}")

# Classifica le rotte come sotto-utilizzate o sovra-utilizzate usando .loc per evitare warning
map_data.loc[:, 'Usage_Category'] = map_data['Seat_Utilization'].apply(
    lambda x: 'Sotto-Utilizzata' if x < global_seat_utilization else 'Sovra-Utilizzata'
)

# Rimuovi duplicati per ogni rotta (solo una riga per ogni combinazione di origin e destination)
map_data_unique = map_data.drop_duplicates(subset=['Origin_airport', 'Destination_airport'])

# Salva il CSV con le 1.000 rotte uniche, coordinate e l'utilizzo dei posti a sedere
map_data_unique.to_csv("Top_1000_Routes_With_Coordinates_And_Utilization_Unique.csv", index=False)
print("CSV con le 1.000 rotte uniche, coordinate e utilizzo dei posti a sedere salvato come 'Top_1000_Routes_With_Coordinates_And_Utilization_Unique.csv'")
```

```python
import pandas as pd
import folium

# Carica il CSV con le rotte uniche, le coordinate e l'utilizzo dei posti a sedere
map_data = pd.read_csv("Top_1000_Routes_With_Coordinates_And_Utilization_Unique.csv")

# Funzione per creare la mappa interattiva con linee proporzionali al traffico passeggeri
def create_flight_map(data, top_n=None):
    """
    Crea una mappa interattiva delle rotte di volo.
    
    Args:
        data: DataFrame con colonne 'Org_airport_lat', 'Org_airport_long', 
              'Dest_airport_lat', 'Dest_airport_long', 'Total_Passengers'.
        top_n: Numero di rotte da visualizzare. Se None, visualizza tutte.
    
    Returns:
        m: Oggetto folium Map.
    """
    # Seleziona le top rotte più trafficate (se specificato)
    if top_n:
        data = data.sort_values(by='Total_Passengers', ascending=False).head(top_n)
    
    # Crea la mappa centrata sulla posizione media degli aeroporti
    avg_lat = data[['Org_airport_lat', 'Dest_airport_lat']].mean().mean()
    avg_long = data[['Org_airport_long', 'Dest_airport_long']].mean().mean()
    m = folium.Map(location=[avg_lat, avg_long], zoom_start=5)
    
    # Calcolo del peso proporzionale in base al traffico passeggeri
    max_passengers = data['Total_Passengers'].max()  # Trova il massimo dei passeggeri
    
    # Aggiungi linee per ogni rotta
    for _, row in data.iterrows():
        # Calcola il peso della linea in base al numero di passeggeri
        weight = (row['Total_Passengers'] / max_passengers) * 10  # Mappa il traffico a uno spessore massimo di 10
        weight = max(1, weight)  # Assicurati che lo spessore non sia mai inferiore a 1
        
        folium.PolyLine(
            locations=[
                [row['Org_airport_lat'], row['Org_airport_long']],
                [row['Dest_airport_lat'], row['Dest_airport_long']]
            ],
            color="blue",
            weight=weight,  # Spessore della linea proporzionale ai passeggeri
            opacity=0.7
        ).add_to(m)
    
    return m

# Crea la mappa per le prime 1.000 rotte più trafficate
top_routes_map = create_flight_map(map_data, top_n=1000)

# Salva la mappa in formato HTML
top_routes_map.save("Top_1000_Routes_Map.html")
print("Mappa interattiva salvata come 'Top_1000_Routes_Map.html'")
```
